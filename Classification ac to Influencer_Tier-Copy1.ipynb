# Univariate Analysis

```python
#taking all the columns where values are continous
contdf = df[['Q_Score','Followers','Avg_Likes','Avg_Views','Engagement_Ratio']]
```

```python
from scipy.stats import skew,kurtosis
key =[]
skewval =[]
kurtval =[]
for i  in contdf:
    print('The skewness of the ',i, '=',skew(contdf[i]))
    print('The kurtosis of the ',i, '=',kurtosis(contdf[i]))
    plt.figure()
    sns.distplot(contdf[i],color='blue',kde=True,
                 hist_kws = {'color':'green', 'edgecolor':'black','linewidth':2,'alpha':0.9})
    plt.show()
    key.append(i)
    skewval.append(skew(contdf[i]))
    kurtval.append(kurtosis(contdf[i]))
```

```python
from scipy.stats import shapiro
shapiro(contdf)
```

**Since the p-value is 0.0 which is less than the alpha(0.5) then we reject the null hypothesis i.e. we have sufficient evidence to say that sample does not come from a normal distribution which is also pretty clear from the above graphs**

```python
dict = {'variables':key,'skew_values':skewval,'kurtosis_values':kurtval}
```

```python
show_val = pd.DataFrame(dict)
```

The rule of thumb for skewness seems to be:

- If the skewness is between -0.5 and 0.5, the data are fairly symmetrical
- If the skewness is between -1 and â€“ 0.5 or between 0.5 and 1, the data are moderately skewed
- If the skewness is less than -1 or greater than 1, the data are highly skewed
- If Skewness is less than 0, then it is called left-skewed or that the left tail is longer than the right tail.
- If Skewness is more than 0, then it is called right-skewed or that the right tail is longer than the left tail.

- the kurtosis of a normal distribution is 0
- So, if a dataset has a positive kurtosis, it has more in the tails than the normal distribution.  If a dataset has a negative kurtosis, it has less in the tails than the normal distribution.

```python
#defining dependent and independent variables
X = df1.drop('Influencer_Tier', axis=1)
y = df1['Influencer_Tier']
```

```python
#splitting data into training and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 1)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
ms = MinMaxScaler()
df[['Followers','Avg_Likes','Avg_Views']] = ms.fit_transform(df[['Followers','Avg_Likes','Avg_Views']])
```

```python
#training model
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,y_train)

#getting confusion matrix
from sklearn.metrics import confusion_matrix,classification_report
y_pred = lr.predict(X_test)
cm = confusion_matrix(y_test,y_pred)
print('confusion matrix:\n',cm)

#checking accuracy
from sklearn.metrics import accuracy_score
lr_acc = accuracy_score(y_test,y_pred)
print('\nAccuracy score = ',round(lr_acc*100,2),'%')
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion = 'entropy')
dt.fit(X_train,y_train)

#getting confusion matrix
from sklearn.metrics import confusion_matrix,classification_report
y_pred = dt.predict(X_test)
cm = confusion_matrix(y_test,y_pred)
print('confusion matrix:\n',cm)

#checking accuracy
from sklearn.metrics import accuracy_score
dt_acc = accuracy_score(y_test,y_pred)
print('\nAccuracy score = ',round(dt_acc*100,2),'%')
```

```python
#training model
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 60, criterion = 'entropy',random_state = 1)
rf.fit(X_train,y_train)

#getting confusion matrix
from sklearn.metrics import confusion_matrix,classification_report
y_pred = rf.predict(X_test)
cm = confusion_matrix(y_test,y_pred)
print('confusion matrix:\n',cm)

#checking accuracy
from sklearn.metrics import accuracy_score
rf_acc = accuracy_score(y_test,y_pred)
print('\nAccuracy score = ',round(rf_acc*100,2),'%')
```

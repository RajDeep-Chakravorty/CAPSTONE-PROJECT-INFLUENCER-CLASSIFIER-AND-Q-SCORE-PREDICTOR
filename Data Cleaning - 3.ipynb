```python
#the k & m is replaced by its actual value i.e thousands and millions respectively
df["Followers"]=df["Followers"].replace({"k":"*1e3", "m":"*1e6"}, regex=True).map(pd.eval).astype(int)
df["Avg_Likes"]=df["Avg_Likes"].replace({"k":"*1e3", "m":"*1e6"}, regex=True).map(pd.eval).astype(int)
df["Avg_Views"]=df["Avg_Views"].replace({"k":"*1e3", "m":"*1e6"}, regex=True).map(pd.eval).astype(int)
```

```python
#Create a column Influencer_Tier according to the number of followers
def cat_followers(x):
    if x>20000 and x<100000:
        return "Micro"
    elif x>100000 and x<500000:
        return "Mid Tier"
    elif x>500000 and x<5000000:
        return "Mega"
    else:
        return "Celebrity"
df['Influencer_Tier']=df['Followers'].apply(cat_followers)
df
```

# Handling the Missing values(0 in our case)

- First we will handle for the column Avg_Likes

```python
#Displaying all the records containing Avg Likes as 0 and Avg Views is not 0
A=df[(df['Avg_Likes']==0) & (df['Avg_Views']!= 0)]
A                #these are basically the values which we would be handling
```

```python
#Displaying all the records where 0 Avg Views and Avg likes is not 0
B=df[(df['Avg_Likes']!=0) & (df['Avg_Views']!= 0)]
B
```

```python
# We have found out the Ratio of Avg_Views and Avg_Likes for each records
Ratio_1 = B['Avg_Views']/B['Avg_Likes']
Ratio_1
```

```python
#Now as a whole we were interested to find out Avg_Likes were what percentage of Avg_Views
Ratio_1.describe()   #But we concluded this would be much better of if we do it according to the Influencer_Tier
```

## Celebrity

```python
# For Celebrity Tier displaying Non 0 Avg_Likes
A_df=df[(df['Influencer_Tier']=='Celebrity') & (df['Avg_Likes']!= 0)]
A_df
```

```python
df_new = df.loc[:,['Name_of_Influencer','Followers','Avg_Likes','Avg_Views','Categories','Influencer_Tier','Q_Score']]
```

```python
df_new
```

```python
#Label Encoding
# Import label encoder
from sklearn import preprocessing
  
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
  
# Encode labels in the following column.
df_new['Categories']= label_encoder.fit_transform(df_new['Categories'])
df_new['Influencer_Tier']= label_encoder.fit_transform(df_new['Influencer_Tier'])
```

```python
df_new.drop(['Name_of_Influencer'],axis=1,inplace=True)
```

```python
df_new
```

```python
#defining dependent and independent variables
X = df_new.drop('Q_Score', axis=1)
y = df_new['Q_Score']
```

```python
#splitting data into training and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 1)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
ms = MinMaxScaler()
df_new[['Followers','Avg_Likes','Avg_Views']] = ms.fit_transform(df_new[['Followers','Avg_Likes','Avg_Views']])
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X_train,y_train)
y_pred=reg.predict(X_test)
rsquare=reg.score(X_test,y_test)
rsquare*100
```

```python
from sklearn.metrics import mean_squared_error
rmse=mean_squared_error(y_test,y_pred,squared=False)
rmse
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor()
dtr.fit(X_train,y_train)
y_pred=dtr.predict(X_test)
rsquare=dtr.score(X_test,y_test)
rsquare*100
```

```python
from sklearn.metrics import mean_squared_error
rmse=mean_squared_error(y_test,y_pred,squared=False)
rmse
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor()
rfr.fit(X_train,y_train)
y_pred=rfr.predict(X_test)
rsquare=rfr.score(X_test,y_test)
rsquare*100
```

```python
from sklearn.metrics import mean_squared_error
rmse=mean_squared_error(y_test,y_pred,squared=False)
rmse
```

```python
### Create a Pickle file using serialization 
import pickle
pickle_out = open("regressor.pkl","wb")
pickle.dump(rfr, pickle_out)
pickle_out.close()
```
